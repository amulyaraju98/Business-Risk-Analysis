```python
product_df['Selling Price'] = (
    product_df['Selling Price'].astype(str)  # ← Force to string
    .str.replace('$', '')
    .str.replace(',', '')
)

product_df['Selling Price'] = pd.to_numeric(product_df['Selling Price'], errors='coerce')

```

```python
print(f"Price range: ${product_df['Selling Price'].min():,.2f} to ${product_df['Selling Price'].max():,.2f}")
```

```python
print(f"\nBefore: {len(product_df)} products")
print(f"Columns: {len(product_df.columns)}")
```

```python
essential_cols = ['Uniqe Id', 'Product Name', 'Category', 'Selling Price']
product_clean = product_df[essential_cols].copy()
```

```python
product_clean = product_clean.dropna(subset=['Uniqe Id', 'Selling Price'])
```

```python
product_clean = product_clean[product_clean['Selling Price'] > 0]
```

```python
print(f"\nAfter: {len(product_clean)} products")
print(f"Columns kept: {list(product_clean.columns)}")
```

```python
product_clean['price_segment'] = pd.qcut(
    product_clean['Selling Price'],
    q=4,
    labels=['Budget', 'Economy', 'Premium', 'Luxury'],
    duplicates='drop'
)

print("Price segment distribution:")
print(product_clean['price_segment'].value_counts().sort_index())

print("\nPrice range by segment:")
for segment in ['Budget', 'Economy', 'Premium', 'Luxury']:
    segment_data = product_clean[product_clean['price_segment'] == segment]['Selling Price']
    if len(segment_data) > 0:
        print(f"  {segment}: ${segment_data.min():.2f} - ${segment_data.max():.2f}")
```

```python
print("Sample of current (long) category names:")
print(product_clean['Category'].head(5).tolist())
```

```python
def simplify_category(cat):
    if pd.isna(cat):
        return 'Other'
    cat_str = str(cat).lower()
    
    if 'sports' in cat_str or 'outdoor' in cat_str:
        return 'Sports'
    if 'toys' in cat_str or 'games' in cat_str:
        return 'Toys'
    if 'clothing' in cat_str or 'fashion' in cat_str or 'apparel' in cat_str:
        return 'Clothing'
    if 'home' in cat_str or 'furniture' in cat_str or 'kitchen' in cat_str:
        return 'Home'
    if 'electronics' in cat_str or 'appliance' in cat_str or 'computer' in cat_str:
        return 'Electronics'
    if 'book' in cat_str or 'media' in cat_str:
        return 'Books'
    if 'beauty' in cat_str or 'health' in cat_str or 'personal' in cat_str:
        return 'Beauty'
    
    return 'Other'
```

```python
product_clean['category_simple'] = product_clean['Category'].apply(simplify_category)

print("\nSimplified categories:")
print(product_clean['category_simple'].value_counts())
```
